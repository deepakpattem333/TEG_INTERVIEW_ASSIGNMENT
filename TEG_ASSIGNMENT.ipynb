```python
import pandas as pd

df = pd.read_excel(r'D:\test1\work.xlsx')
```

```python
df['out-inn'] = df["out"] - df["inn"]
```

```python
counti=0
for i in df['out-inn']:
    if i<0:
        df['out-inn'][counti]=i+24
        counti=counti+1
    else:
        counti=counti+1
df
```

```python
out_inn=[]
inn=[]
for i in df['out-inn'].iteritems():
    out_inn.append(i[1])
for i in df['inn'].iteritems():
    inn.append(i[1])
 
```

```python
l = [0] * 24

for count,item in enumerate(inn):
    mmp=item
    temp=out_inn[count]
    
    if mmp==24:
        mmp=0
    
    while temp > 0 and mmp<24:
        
        l[mmp]=l[mmp]+1
        
        temp=temp-1
        
        if mmp < 24:
            mmp=mmp+1   
        
        if mmp == 24:    
            mmp = 0

        
```

```python
dict={}
for i in range(24):
    tempo=i+1
    string=str(i)+"-"+str(tempo)
    dict[string]=l[i]
    
```
